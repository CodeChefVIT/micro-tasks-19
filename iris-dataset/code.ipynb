```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#importing data using numpy
#data=np.recfromcsv('dataset.csv')
#importing data using pandas
data1=pd.read_csv('dataset.csv',nrows=None)
data1=data1.drop(columns=['Id'])
print(f'The dataset contains {data1.shape[0]} rows and {data1.shape[1]} columns')
print(data1)
#npdata=data1.values
#print(npdata)
histograms(data1,5)
plottingGraphs(data1.iloc[:,0:-1],4)
correlation(data1.iloc[:,0:-1],5)
sp=list(set(data1['Species']))
for k in sp:
    print(f'For the species {k}:')
    dff=data1.iloc[[i for i in range(data1.shape[0]) if data1.iloc[i,-1]==k],0:data1.shape[1]]
    dff=dff.drop(columns=['Species'])
    print(f'The data contains {dff.shape[0]} rows')
    histograms(dff,5)
    plottingGraphs(dff,4)
    correlation(data1.iloc[:,0:-1],5)
        
#############################################
#commands for visualization
#pd.DataFrame.hist(data1[['PetalWidthCm']])
#plt.show()
#plt.scatter(data1[['PetalWidthCm']],data1[['SepalWidthCm']])
##############################################
```

```python
def histograms(df,nGraphPerRow):
    nr,nc=data1.shape
    df=df.iloc[0:nr,0:nc]
    df=df.dropna('columns')
    nGraphRow=0 if nc%nGraphPerRow==0 else 1
    nGraphRow+=nc/nGraphPerRow
    plt.figure(num = None, figsize = (5 * nGraphPerRow, 5 * nGraphRow), dpi = 80, facecolor = 'w', edgecolor = 'k')
    for i in range(df.shape[1]):
        plt.subplot(nGraphRow, nGraphPerRow, i + 1)
        columnDf = df.iloc[0:nr, i]        
        if (not np.issubdtype(type(columnDf.iloc[0]), np.number)):
            valueCounts = columnDf.value_counts()
            valueCounts.plot.bar()
        else:
            columnDf.hist()
        plt.ylabel('counts')
        plt.xticks(rotation = 90)
        plt.title(f'{list(df)[i]} (column {i})')
    plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
    plt.show()
    
```

```python
def plottingGraphs(df,nGraphPerRow):
    nr,nc=data1.shape
    df=df.iloc[0:nr,0:nc]
    df=df.dropna('columns')
    nGraphRow=0 if (nc**2)%nGraphPerRow==0 else 1
    nGraphRow+=(nc**2)/nGraphPerRow
    plt.figure(num = None, figsize = (5 * nGraphPerRow, 5 * nGraphRow), dpi = 80, facecolor = 'w', edgecolor = 'k')
    x=0
    corrs = df.corr().values
    for i in range(df.shape[1]):
        for j in range(df.shape[1]):
            plt.subplot(nGraphRow, nGraphPerRow,x+1)
            x+=1
            columnDf = df.iloc[0:nr, i]
            column2Df = df.iloc[0:nr, j]
            plt.scatter(columnDf,column2Df)
            plt.xlabel(list(df)[i])
            plt.ylabel(list(df)[j])
            plt.xticks(rotation = 90)
            plt.annotate('Corr. coef = %.3f' % corrs[i, j], (0.8, 0.2), xycoords='axes fraction', ha='center', va='center', size=10)
    plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
    plt.show()
```

```python
def correlation(df,graphWidth):
    df=df.dropna('columns')
    df=df[[col for col in df if df[col].nunique() > 1]]
    if df.shape[1]<2:
        print('co-relation not possible as the number of non-constant(or not containing NaN) columns is less than 2')
        return
    corr = df.corr()
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, fignum = 1)#for the colour plot
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)#to get ticks for the plot
    plt.yticks(range(len(corr.columns)), corr.columns)#to get ticks for the plot
    plt.gca().xaxis.tick_bottom()#to get the ticks to the bottom which are at top by default
    plt.colorbar(corrMat)#for the index
    plt.title('Correlation Matrix', fontsize=15)
    plt.show()
```
